```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Residential_Property_Assessments.csv')

df.head()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Livable Area'], df['Assessment'], alpha=0.6)
plt.xlabel('Livable Area')
plt.ylabel('Assessment')
plt.title('Residential Property Assessments')
```

```python
X = df[['Livable Area']].values
y = df['Assessment'].values

model = LinearRegression()
model.fit(X, y)
intercept = model.intercept_
slope = model.coef_[0]

print(f"Intercept: {intercept:.2f}")
print(f"Slope: {slope:.2f}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Livable Area'], df['Assessment'], alpha=0.6)
plt.plot(X, model.predict(X), color='red', linewidth=1.5, label='Regression Line')
plt.xlabel('Livable Area')
plt.ylabel('Assessment')
plt.title('Residential Property Assessments')
plt.legend()
```

```python
model.predict([[2750]]) 
```

```python
# Show that residuals have mean zero 
residuals = y - model.predict(X) 
print(f"Residuals Mean: {residuals.mean():.2f}")
```

```python
# The predicted values and residuals from this regression are uncorrelated 
pred = model.predict(X) 
print(np.corrcoef(pred, residuals)[0, 1])
```
